```python
import sys
sys.path.insert(1, '..')
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from ProdSeqAnalysis_utils import *
# plt.rcParams['svg.fonttype'] = 'none'
# plt.rcParams["font.family"] = "Arial"
```

```python
DIPG_signalpath = "/path/to/DIPG_output.PPIEnrichment.tsv"
DIPG_signals_unordered, DIPG_nms_unordered, DIPG_pairnms = ReadProdTSVFile(DIPG_signalpath)

reorder_idx = [
    14, 15, 4, 5, 
    18, 19, 8, 9, 
    16, 17, 6, 7, 
    12, 13, 2, 3, 
    10, 11, 0, 1, 
]

DIPG_signals = [DIPG_signals_unordered[x] for x in reorder_idx]

DIPG_signals_normed = []
for row in DIPG_signals:
    DIPG_signals_normed.append([x/row[49] for x in row])

DIPG_nms = [DIPG_nms_unordered[x] for x in reorder_idx]
```

```python
DIPG_signals_normedavg = []
DIPG_nmsavg = []
DIPGavg_idx_list = [2 * x for x in range(int(len(DIPG_nms) / 2))]
for idx in DIPGavg_idx_list:
    curr_row = []
    for j in range(len(DIPG_signals_normed[idx])):
        curr_mean = (DIPG_signals_normed[idx][j] + DIPG_signals_normed[idx + 1][j]) / 2
        curr_row.append(curr_mean)
    DIPG_signals_normedavg.append(curr_row)
    DIPG_nmsavg.append(DIPG_nms[idx])
```

```python

PPI_plot_groups_reorder = [
    ["EED&EZH2", "EED&SUZ12", "EZH2&SUZ12"],
    ["AEBP2&EED", "AEBP2&EZH2", "AEBP2&SUZ12"],
    ["H3K4me3&H3K27ac", "H3K27M&EED", "H3K27M&EZH2", "H3K27M&SUZ12", "H3K27M&H3K27ac"],
    ['H3K27me3&EED', 'H3K27me3&EZH2', "H3K27me3&SUZ12"], 
    ["H3K27ac&EED"]
]

plot_cmaps = [
    sns.color_palette("YlOrBr", as_cmap = True),
    sns.color_palette("Purples", as_cmap = True),
    sns.color_palette("Greens", as_cmap = True),
    sns.color_palette("Blues", as_cmap = True),
    sns.color_palette("pink_r", as_cmap = True),
]

fig = GroupedPPIHeatmap(DIPG_signals_normedavg[0:2], DIPG_nmsavg[0:2], DIPG_pairnms, bc_pair_groups = PPI_plot_groups_reorder,
                  plot_cmaps = plot_cmaps,
                  sample_group_sizes = [2, 2, 2, 2, 2], plot_vmax = [10, 1, 2, 1, 0.6],
                  cbar_ax_locs = [
                      [0.13, 0.88, 0.15, 0.015],
                      [0.353, 0.88, 0.15, 0.015],
                      [0.578, 0.88, 0.15, 0.015],
                      [0.798, 0.88, 0.05, 0.015],
                      [0.9, 0.88, 0.05, 0.015],
                  ],
                  plot_size = (12, 2)
                 )
# fig.savefig("DIPG38_heatmap.svg", dpi = 1200)
```
