```python
example_qc_path = "/path/to/DOL_output.QCCnts.tsv"
example_qc_lines = [x.strip().split("\t") for x in open(example_qc_path).readlines()]
example_qc_nms = [x[0] for x in example_qc_lines[1:]]
example_qc_cnts = [[float(y) for y in x[1:]] for x in example_qc_lines[1:]]
```

```python
example_qc_selfself_cnt = [sum(x[3:]) for x in example_qc_cnts]
example_qc_selfself_prop = [example_qc_selfself_cnt[idx] / example_qc_cnts[idx][2] for idx in range(len(example_qc_cnts))]
example_qc_nonselfself_UMIprop = [1-x for x in example_qc_selfself_prop]
# example_qc_nonselfself_UMIprop.extend([
#     105578 / 723785,
#     89169 / 543442,
#     99735 / 791716,
#     107225 / 568918,
#     107495 / 745889,
# ])
```
