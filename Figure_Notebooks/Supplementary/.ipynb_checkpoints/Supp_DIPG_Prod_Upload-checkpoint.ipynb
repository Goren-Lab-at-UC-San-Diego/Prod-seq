```python
import sys
sys.path.insert(1, '../..')
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
from ProdSeqAnalysis_utils import *
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams["font.family"] = "Arial"
from matplotlib.patches import ConnectionPatch
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
```

```python
DIPG_signalpath = "/path/to/SUDIPG_output.PPIEnrichment.tsv"
DIPG_signals_unordered, DIPG_nms_unordered, DIPG_pairnms = ReadProdTSVFile(DIPG_signalpath)

reorder_idx = [
    15, 14, 5, 4,
    18, 18, 9, 8,
    17, 16, 7, 6,
    13, 12, 3, 2,
    11, 10, 1, 0,
]

DIPG_signals = [DIPG_signals_unordered[x] for x in reorder_idx]

DIPG_signals_normed = []
for row in DIPG_signals:
    DIPG_signals_normed.append([x/row[49] for x in row])

DIPG_nms = [DIPG_nms_unordered[x] for x in reorder_idx]
```

```python
PPI_plot_groups_reorder = [
    ["EED&EZH2", "EED&SUZ12", "EZH2&SUZ12"],
    ["AEBP2&EED", "AEBP2&EZH2", "AEBP2&SUZ12"],
    ["H3K4me3&H3K27ac", "H3K27M&EED", "H3K27M&EZH2", "H3K27M&SUZ12", "H3K27M&H3K27ac"],
    ['H3K27me3&EED', 'H3K27me3&EZH2', "H3K27me3&SUZ12"], 
    ["H3K27ac&EED"]
]

plot_cmaps = [
    sns.color_palette("YlOrBr", as_cmap = True),
    sns.color_palette("Purples", as_cmap = True),
    sns.color_palette("Greens", as_cmap = True),
    sns.color_palette("Blues", as_cmap = True),
    sns.color_palette("pink_r", as_cmap = True),
]

fig = GroupedPPIHeatmap(DIPG_signals_normed[0:4], DIPG_nms[0:4], DIPG_pairnms, bc_pair_groups = PPI_plot_groups_reorder,
                  plot_cmaps = plot_cmaps,
                  sample_group_sizes = [4, 4, 4, 4, 4], plot_vmax = [10, 1, 2, 1, 0.6],
                  cbar_ax_locs = [
                      [0.13, 0.88, 0.15, 0.015],
                      [0.353, 0.88, 0.15, 0.015],
                      [0.578, 0.88, 0.15, 0.015],
                      [0.798, 0.88, 0.05, 0.015],
                      [0.9, 0.88, 0.05, 0.015],
                  ],
                  plot_size = (12, 2)
                 )
# fig.savefig("DIPG38_unmerge_heatmap.svg", dpi = 1200)
```

```python
otheroption_bio1_signalpath = "/path/to/SUDIPG_bio1_output.PPIEnrichment.tsv"
otheroption_bio1_signals_unorder, otheroption_bio1_nms_unorder, otheroption_bio1_pairnms = ReadProdTSVFile(otheroption_bio1_signalpath)

otheroption_bio2_signalpath = "/path/to/SUDIPG_bio2_output.PPIEnrichment.tsv"
otheroption_bio2_signals_unorder, otheroption_bio2_nms_unorder, otheroption_bio2_pairnms = ReadProdTSVFile(otheroption_bio2_signalpath)


reorder_idx = [
    4,
    8,
    6,
    2,
    0,
]
otheroption_nms = []
otheroption_signals_normed = []
for idx in reorder_idx:
    otheroption_signals_normed.append([x / otheroption_bio2_signals_unorder[idx + 1][49] for x in otheroption_bio2_signals_unorder[idx + 1]])
    otheroption_signals_normed.append([x / otheroption_bio2_signals_unorder[idx][49] for x in otheroption_bio2_signals_unorder[idx]])
    otheroption_signals_normed.append([x / otheroption_bio1_signals_unorder[idx + 1][49] for x in otheroption_bio1_signals_unorder[idx + 1]])
    otheroption_signals_normed.append([x / otheroption_bio1_signals_unorder[idx][49] for x in otheroption_bio1_signals_unorder[idx]])
    
    otheroption_nms.append(otheroption_bio2_nms_unorder[idx + 1])
    otheroption_nms.append(otheroption_bio2_nms_unorder[idx])
    otheroption_nms.append(otheroption_bio1_nms_unorder[idx + 1])
    otheroption_nms.append(otheroption_bio1_nms_unorder[idx])
    
```
