```python
import sys
sys.path.insert(1, '../..')
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
from ProdSeqAnalysis_utils import *
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams["font.family"] = "Arial"
from matplotlib.patches import ConnectionPatch
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
```

```python
washtimes_bio1_signalpath = "/path/to/HeLa_bio1_output.PPIEnrichment.tsv"
washtimes_bio1_signals, washtimes_bio1_nms, washtimes_bio1_pairnms = ReadProdTSVFile(washtimes_bio1_signalpath)

washtimes_bio2_signalpath = "./path/to/HeLa_bio2_output.PPIEnrichment.tsv"
washtimes_bio2_signals, washtimes_bio2_nms, washtimes_bio2_pairnms = ReadProdTSVFile(washtimes_bio2_signalpath)

idx_list = [
    3, 2,
    1, 0,
    11, 10,
    9, 8,
]
washtimes_signals_normed = []
washtimes_nms = []
washtimes_pairnms = washtimes_bio1_pairnms
for idx in idx_list:
    washtimes_signals_normed.append([x / (washtimes_bio2_signals[idx + 4][49]) for x in washtimes_bio2_signals[idx + 4]])
    washtimes_signals_normed.append([x / (washtimes_bio2_signals[idx][49]) for x in washtimes_bio2_signals[idx]])

    washtimes_signals_normed.append([x / (washtimes_bio1_signals[idx + 4][49]) for x in washtimes_bio1_signals[idx + 4]])
    washtimes_signals_normed.append([x / (washtimes_bio1_signals[idx][49]) for x in washtimes_bio1_signals[idx]])
    

    washtimes_nms.append(washtimes_bio2_nms[idx + 4])
    washtimes_nms.append(washtimes_bio2_nms[idx])

    washtimes_nms.append(washtimes_bio1_nms[idx + 4])
    washtimes_nms.append(washtimes_bio1_nms[idx])
    
    
```

```python
plot_cmaps = [
    sns.color_palette("YlOrBr", as_cmap = True),
    sns.color_palette("Purples", as_cmap = True),
    sns.color_palette("Greens", as_cmap = True),
    sns.color_palette("Blues", as_cmap = True),
]

PPI_plot_groups = [
    ["EED&EZH2", "EED&SUZ12", "EZH2&SUZ12"],
    ["AEBP2&EED", "AEBP2&EZH2", "AEBP2&SUZ12", "H3K4me3&H3K27ac"],
    ["H3K27M&EED", "H3K27M&EZH2", "H3K27M&SUZ12", "H3K27M&H3K27ac"]
]

fig = GroupedPPIHeatmap(washtimes_signals_normed[0:16], washtimes_nms[0:16], washtimes_pairnms, bc_pair_groups = PPI_plot_groups,
                  plot_cmaps = plot_cmaps,
                  sample_group_sizes = [4, 4, 4, 4, 4, 4, 4, 4], plot_vmax = [6.5, 1.5, 2],
                        plot_vmin = [0, 0, 0],
                  cbar_ax_locs = [
                      [0.161, 0.87, 0.15, 0.01],
                      [0.37, 0.87, 0.15, 0.01],
                      [0.62, 0.87, 0.15, 0.01],
                      [0.845, 0.87, 0.05, 0.01],
                  ],
                  plot_size = (10, 8)
                 )
# fig.savefig("washouttimes_unmerge_left.svg", dpi = 1200)
```
