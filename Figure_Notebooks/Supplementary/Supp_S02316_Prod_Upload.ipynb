```python
import sys
sys.path.insert(1, '../..')
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
from ProdSeqAnalysis_utils import *
# plt.rcParams['svg.fonttype'] = 'none'
# plt.rcParams["font.family"] = "Arial"
from matplotlib.patches import ConnectionPatch
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
```

```python
S16bio1_signalpath = "/path/to/S02316_output_bio1.PPIEnrichment.tsv"
S16bio1_signals, S16bio1_nms, S16bio1_pairnms = ReadProdTSVFile(S16bio1_signalpath)

S16bio2_signalpath = "/path/to/S02316_output_bio2.PPIEnrichment.tsv"
S16bio2_signals, S16bio2_nms, S16bio2_pairnms = ReadProdTSVFile(S16bio2_signalpath)

S16_signals_normed = []
S16_nms = []

idx_list = [3, 2, 1, 0]
for i in idx_list:
    S16_signals_normed.append([x/(S16bio2_signals[2 * i + 1])[49] for x in S16bio2_signals[2 * i + 1]])
    S16_signals_normed.append([x/(S16bio2_signals[2 * i])[49] for x in S16bio2_signals[2 * i]])
    S16_signals_normed.append([x/(S16bio1_signals[2 * i + 1])[49] for x in S16bio1_signals[2 * i + 1]])
    S16_signals_normed.append([x/(S16bio1_signals[2 * i])[49] for x in S16bio1_signals[2 * i]])
    
    S16_nms.append(S16bio2_nms[2 * i + 1])
    S16_nms.append(S16bio2_nms[2 * i])
    S16_nms.append(S16bio1_nms[2 * i + 1])
    S16_nms.append(S16bio1_nms[2 * i])


S16_pairnms = S16bio2_pairnms
```

```python
plot_cmaps = [
    sns.color_palette("YlOrBr", as_cmap = True),
    sns.color_palette("Purples", as_cmap = True),
    sns.color_palette("Greens", as_cmap = True),
    sns.color_palette("Blues", as_cmap = True),
]

PPI_plot_groups = [
    ["EED&EZH2", "EED&SUZ12", "EZH2&SUZ12"],
    ["AEBP2&EED", "AEBP2&EZH2", "AEBP2&SUZ12", "H3K4me3&H3K27ac"],
    ["H3K27M&EED", "H3K27M&EZH2", "H3K27M&SUZ12", "H3K27M&H3K27ac"]
]

fig = GroupedPPIHeatmap(S16_signals_normed[0:8], S16_nms[0:8], S16_pairnms, bc_pair_groups = PPI_plot_groups,
                  plot_cmaps = plot_cmaps,
                  sample_group_sizes = [4, 4, 4, 4], plot_vmax = [6.5, 3, 2],
                        plot_vmin = [0, 0, 0],
                  cbar_ax_locs = [
                      [0.161, 0.87, 0.15, 0.01],
                      [0.37, 0.87, 0.15, 0.01],
                      [0.62, 0.87, 0.15, 0.01],
                      [0.845, 0.87, 0.05, 0.01],
                  ],
                  plot_size = (10, 6)
                 )

# fig.savefig("S02316_unmerge_left.svg", dpi = 1200)
```
